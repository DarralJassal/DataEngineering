```python
#Data cleaning - dropping columns that contain null values
dfschools = data_schools.dropna(axis=1)
dfschools.head()
```

```python
#dropping all columns with null values in tables for charities
dfcharities = data_charities.dropna(axis=1)
dfcharities
```

```python
count=dfcharities['charity_insolvent'].value_counts()
count1=dfcharities['charity_in_administration'].value_counts()
count1
```

```python
#removing charities that are dissolved
dfcharities1 = dfcharities.loc[~dfcharities.charity_insolvent, :]
```

```python
#checking that charities which are dissolved have been removed
count=dfcharities1['charity_insolvent'].value_counts()
count
```

```python
dfcharities1 = dfcharities.loc[~dfcharities.charity_in_administration, :]
```

```python
count1=dfcharities1['charity_in_administration'].value_counts()
count1
```

```python
#dropping charities which have had registriation removed.
dfcharities1 = dfcharities1[dfcharities1['charity_registration_status'].str.contains('Registered')]
dfcharities1
```

```python
dfcharities2 = dfcharities1.drop('charity_registration_status',1)
dfcharities3 = dfcharities2.drop('charity_insolvent',1)
dfcharities4 = dfcharities3.drop('charity_in_administration',1)
dfcharities5 = dfcharities4.drop('date_of_extract',1)
dfcharities5
```

```python
#final table for charities
dfcharities=dfcharities5
dfcharities
```
